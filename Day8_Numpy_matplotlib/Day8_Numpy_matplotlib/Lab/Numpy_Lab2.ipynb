```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot  as plt
%matplotlib inline

import utils.image_audio as ia
from utils.checkarr import array_hash, check_hash
from utils.float_inspector import print_float, raw_binary_view, binary_to_array, get_binary
```

```python
models = np.loadtxt("Data/finmodel.txt")
```

```python
current_value = 1535135353000.99
```

```python
#take the culmulative sum of model predicitons over rows
cumulative_gains = np.cumsum(models, axis=0)
```

```python
value_of_company = current_value + cumulative_gains
#standard deviation over columns to estimate how closely the ensemble of models match each other at each time point
std_dev = np.std(value_of_company, axis=1)
```

```python
# plot the results
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(std_dev)
ax.set_xlabel("Time (minutes)")
ax.set_ylabel("Variation in value across models")
```
